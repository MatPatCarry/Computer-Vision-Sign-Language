```python
import numpy as np

f = np.zeros(shape=(1, 12, 3))
ar = np.random.uniform(size=(23, 12, 3))
l = np.zeros(shape=(1, 12, 3))

ar = np.concatenate((f, ar, l), axis=0)
```

```python
h = 'dd dssfd aa'.capitalize()
h
```

```python
all_zeros_mask = np.all(ar == 0, axis=(1, 2))
```

```python
result = np.diff(ar, axis=0)
```

```python
for i in range(len(result)):
    
    if any(all_zeros_mask[i: i + 2]): 
        result[i, :, :] = 0
```
